```python
import pickle
import matplotlib.pyplot as plt
```

```python
def plot_lc(loss_dict_name, title):
    with open(f'{loss_dict_name}.pickle', 'rb') as handle:
        loss_dict = pickle.load(handle) 
        plt.title(title)
        plt.plot(loss_dict['train'], label = f'{loss_dict_name} train')#, color = 'black')
        plt.plot(loss_dict['val'], label = f'{loss_dict_name} validation')#, color = 'royalblue')
        plt.xlabel('Number of epochs')
        plt.ylabel('MSE')
        plt.legend()   
        
        try:
            print(loss_dict['maximum_distance'])
        except:
            print('no att')
```

```python
plot_lc('fcn', 'Fully Connected')
```

```python
plot_lc('gcn', 'Graph ConvNet')
```

```python
plot_lc('gcn_concat', 'Graph ConvNet - concat')
plot_lc('gcn_sum', 'Graph ConvNet - sum')
plot_lc('fcn', 'Fully Connected')
```
